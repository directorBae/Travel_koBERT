```python
def get_label(df):
  labels = []
  for i in range(len(df)):
    temp = {}
    for word in word_dict['words'].tolist():
      temp.setdefault(word, 0)
    for token in df['token'].iloc[i]:
      for word in word_dict['words'].tolist():
        try:
          temp[word] += TLD_wv.wv.similarity(str(word), str(token))
        except:
          pass
    dict_sorted = sorted(temp.items(), key=lambda x: x[1], reverse=True)
    labels.append(dict_sorted[0:10])
    print("\rProgress: "+str((i+1)/len(df)*100)+"%          ", end='')
  return labels
```
